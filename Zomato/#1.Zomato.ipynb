```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx' )
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how='left')
```

```python
## Check Data Type

final_df.dtypes
```

```python
final_df
```

```python
country_values = final_df.Country.value_counts().values
country_names = final_df.Country.values
```

```python
## Pie chart
```

```python
plt.pie(country_values[:3], labels=country_names[:3], autopct="%1.2f%%")
```

```python
## Observations: Maximum transactions are from India
```

```python
ratings = final_df.groupby(["Aggregate rating", 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
```

```python
ratings
```

## Conclusions

1. Rating between 4.5 and 4.9 --> Excellent
2. Rating between 4.0 and 4.4 --> Very good
3. Rating between 3.5 and 3.9 --> Good
4. Rating between 2.5 and 3.4 --> Average
5. Rating between 1.8 and 2.4 --> Poor
6. Rating is 0 --> Not rated



```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observations:
1. Not rated count is very high
2. Max. # of ratings are b/w 2.8 and 3.9

```python
## Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
